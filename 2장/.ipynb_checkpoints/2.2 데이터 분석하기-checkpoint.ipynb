# 2.2.1 모멘텀 지표 분석하기

```python
import numpy as np
import FinanceDataReader as fdr

def get_momentum(scode, SD, ED):
    """    
    주가 모멘텀과 거래량 모멘텀 계산 

    매개변수: 
         scode : 종목 코드 (예: '005930')
         SD : 주가데이터 추출시작일 (예: '20250102')
         ED : 주가데이터 추출종료일 (예: '20250131')

    반환값: 
        avg : 주가 모멘텀 지표 (기간별 수익률 평균)
        vol : 거래량 모멘텀 지표 
    """

    # 1. 주가 및 거래량 데이터 수집 (여기는 에러나면 멈춰서 확인해야 하므로 try-except 제거)
    df = fdr.DataReader(scode, SD, ED)

    # 2. 주가 모멘텀 지표 생성
    # 단기 (1일, 1주, 1개월) - 데이터가 너무 없으면 에러가 나는 게 맞으므로 그대로 둠
    mom_1d = ( df.iloc[-1]['Close'] / df.iloc[-2]['Close'] - 1 )  * 100    # 전일
    mom_1w = ( df.iloc[-1]['Close'] / df.iloc[-5]['Close'] - 1 )  * 100    # 주간
    mom_1m = ( df.iloc[-1]['Close'] / df.iloc[-20]['Close'] - 1 )  * 100   # 월간 

    # 3개월간 (데이터 부족 시 NaN 처리)
    try:
        mom_3m = ( df.iloc[-1]['Close'] / df.iloc[-60]['Close'] - 1 )  * 100 
    except:
        mom_3m = np.nan

    # 연간 (데이터 부족 시 NaN 처리)
    try:
        mom_1y = ( df.iloc[-1]['Close'] / df.iloc[-240]['Close'] - 1 )  * 100
    except:
        mom_1y = np.nan

    # 3. 기간별 수익률의 평균 계산 (NaN을 제외하고 평균 산출)
    avg = np.nanmean([mom_1d, mom_1w, mom_1m, mom_3m, mom_1y])

    # 4. 거래량 모멘텀 지표 생성
    try:
        vol_short = df['Volume'].rolling(window=5).mean().iloc[-1]  # 최근 5일
        vol_long = df['Volume'].rolling(window=20).mean().iloc[-1]  # 최근 20일
        vol = ( vol_short / vol_long - 1) * 100
    except:
        vol = np.nan

    return avg, vol
```

```python
fdr.DataReader('005930','20240920', '20250930') # 삼성전자
```

```python
# (참고) 코스피200 종목코드 추출 
df = ( fdr.StockListing('KOSPI')[['Code','Name','Marcap']]
          .query("Name not in ['삼성전자우', '현대차2우B','미래에셋증권2우B','삼성에피스홀딩스','맥쿼리인프라']")
          .iloc[:200]
          .reset_index(drop=True) )

stock_list = dict(zip(df['Code'], df['Name']))
```

```python
# 코스피200 종목에 대해서 모멘텀 지표 계산 후 result에 적재  
result = []
for scode, sname in stock_list.items():
    mom, vol = get_momentum(scode, '20240920', '20250930')  # 최근 1년 
    result.append( [ scode, sname, mom, vol ] )

# 모멘텀 지표 결과(result)는 데이터프레임(data_mom)으로 저장
data_mom = pd.DataFrame(result, columns=['scode', 'sname', 'mom_price', 'mom_vol'])

# 마지막 5건 출력
data_mom.tail()
```

```python
data_mom.describe()
```

```python
# 코스피 200 12월 일별 주가 데이터 추출 
kospi200 = ','.join( stock_list.keys() )
data = fdr.DataReader(kospi200, '20251001', '20251230')

first_row = data.iloc[0]  # 10월 1일
last_row = data.iloc[-1]  # 12월 30일
ret = ((last_row - first_row) / first_row) * 100

df_ret = pd.DataFrame({'scode':ret.index, 'RET':ret.values})
df_ret
```

```python
# 모멘텀 지표와 결합. how='left'옵션으로 모멘텀 지표 기준 조인 
df = data_mom.merge(df_ret, on='scode', how='left')

# 최종 3건 화면 출력
df.tail(3)
```

```python
import matplotlib.pyplot as plt
def plot_by_factor(df, feat):
    ser = pd.qcut(df[feat], 10, duplicates='drop')                   # feat 컬럼별 10개 그룹화 
    print( df.groupby(ser)['RET'].agg(['size', 'mean', 'median']) )  # 그룹별 건수, 평균, 중앙값 출력 
    df.groupby(ser)['RET'].mean().plot(kind='bar')                   # 그룹별 수익률 평균으로 막대 그래프 생성
    plt.show()
```

```python
plot_by_factor(df, 'mom_price')
```

```python
plot_by_factor(df, 'mom_vol')
```

# 2.2.2 밸류 지표 분석하기

```python
import FinanceDataReader as fdr
import pandas as pd

target_stock = "005930" # 삼성전자
# 재무제표 데이터 추출 (삼성전자, 연간 K-IFRS 연결 기준)
data = fdr.SnapDataReader(f'NAVER/FINSTATE-2Y/{target_stock}') 

# 미래 추정치 제외 (영업이익 결측치 삭제) 
data_val_raw = data.dropna(subset='영업이익')
data_val_raw['scode'] = target_stock
# PER, PBR 출력 
data_val = data_val_raw[['scode', 'PER(배)', 'PBR(배)']]\
            .reset_index()\
            .rename(columns={'PER(배)':'PER','PBR(배)':'PBR'})

# 최종 데이터를 화면에 출력   
data_val.tail()
```

```python
# 시가총액 상위 200개 종목 추출 (Top 200)
df = ( fdr.StockListing('KOSPI')[['Code','Name','Marcap']]
          .query("Name not in ['삼성전자우', '현대차2우B','미래에셋증권2우B','삼성에피스홀딩스','맥쿼리인프라']")
          .iloc[:200]
          .reset_index(drop=True) )

stock_list = dict(zip(df['Code'], df['Name']))
stock_list

# 200개 종목에 대한 데이터 추출 
result = []
for scode, sname in stock_list.items():
    try: 
        data = fdr.SnapDataReader(f'NAVER/FINSTATE-2Y/{scode}').dropna(subset='영업이익')
        per, pbr = data['PER(배)'].iloc[-1], data['PBR(배)'].iloc[-1]
        result.append([scode, per, pbr])

    except: 
        print("추출오류 : ", scode) # 에러 종목코드 출력
        result.append([scode, None, None])

df = pd.DataFrame(result, columns=['scode', 'PER', 'PBR'])
df.head()
```

```python
df.describe() 
```

```python
# 주가 데이터 추출  
# 코스피 200 12월 일별 주가 데이터 추출 
kospi200 = ','.join( stock_list.keys() )
data = fdr.DataReader(kospi200, '20251001', '20251230')

first_row = data.iloc[0]  # 12월 1일
last_row = data.iloc[-1]  # 12월 30일
ret = ((last_row - first_row) / first_row) * 100

df_ret = pd.DataFrame({'scode':ret.index, 'RET':ret.values})

# 밸류 데이터와 결합 
data_val = df.merge(df_ret, on='scode') 
data_val

# 최종 3건 화면 출력 
data_val.head(3)
```

```python
import matplotlib.pyplot as plt
def plot_by_factor(df, feat):
    ser = pd.qcut(df[feat], 10, duplicates='drop')                   # feat 컬럼별 10개 그룹화 
    print( df.groupby(ser)['RET'].agg(['size', 'mean', 'median']) )  # 그룹별 건수, 평균, 중앙값 출력 
    df.groupby(ser)['RET'].mean().plot(kind='bar')                   # 그룹별 수익률 평균으로 막대 그래프 생성
    plt.show()

plot_by_factor(data_val, 'PER')
```

```python
plot_by_factor(data_val, 'PBR')
```

# 2.2.3 퀄러티 지표 분석하기

```python
import FinanceDataReader as fdr
import pandas as pd
import numpy as np

# 대상 컬럼 지정 
cols = ['매출액', '영업이익', '당기순이익', 'ROE(%)'] 

# 재무 데이터 추출 : 분기 K-IFRS 연결 기준 
data = fdr.SnapDataReader('NAVER/FINSTATE-2Q/005930')[cols] 

# 미래 추정치 제외 : 영업이익 결측치 삭제 
data = data.dropna(subset='영업이익')

# 결과 화면 출력 
data
```

```python
def get_quality(scodes):
    # 결과를 저장할 빈 리스트 생성
    result = []

    # scodes의 종목코드 컬럼을 하나씩 반복
    for scode in scodes:
        try: 
            # 1. 데이터 수집 : 해당 종목의 재무제표를 가져옴 
            #  - dropna(): '영업이익' 데이터 결측 행 제거
            data = fdr.SnapDataReader(f'NAVER/FINSTATE-2Q/{scode}').dropna(subset='영업이익')

            # 2.재무데이터 추출 : 매출액, 영업이익, 당기순이익 증가율 (최근 분기) 
            fin = (data[['매출액','영업이익','당기순이익']].pct_change()*100).iloc[-1].to_list()

            # 3. ROE 추출 : 최근 시점
            roe = data['ROE(%)'].iloc[-1]

            # 4. 데이터 저장: [종목코드, 매출액증가율 영업이익증가율, 당기순이익증가율, ROE] 형태로 result에 추가
            result.append([scode] + fin + [roe])

        except: 
            # 4. try 블록 실행 중 에러 발생시 실행
            print("추출오류 : ", scode) # 에러 종목코드 출력

            # 에러 종목별 'None' 채워서 result에 추가
            result.append([scode, None, None, None, None])

    # result를 데이터프레임으로 변환 
    df = pd.DataFrame(result, columns=['scode', 'revenue_rate', 'oper_income_rate', 'net_income_rate', 'ROE'])

    # 결과 화면 출력 
    return df
```

```python
## 코스피  200 종목 리스트 산출
df = ( fdr.StockListing('KOSPI')[['Code','Name','Marcap']]
          .query("Name not in ['삼성전자우', '현대차2우B','미래에셋증권2우B','삼성에피스홀딩스','맥쿼리인프라']")
          .iloc[:200]
          .reset_index(drop=True) )

stock_list = dict(zip(df['Code'], df['Name']))
stock_list
```

```python
data_fin = get_quality(stock_list.keys())
data_fin.head()
```

```python
data_fin. describe()
```

```python
# 검증용 주가 데이터 추출 
# 코스피 200 12월 일별 주가 데이터 추출 
kospi200 = ','.join( stock_list.keys() )
data = fdr.DataReader(kospi200, '20251001', '20251230')

first_row = data.iloc[0]  # 10월 1일
last_row = data.iloc[-1]  # 12월 30일
ret = ((last_row - first_row) / first_row) * 100

df_ret = pd.DataFrame({'scode':ret.index, 'RET':ret.values})


# 모멘텀 지표와 결합. how='left'옵션으로 모멘텀 지표 기준 조인 
df = data_fin.merge(df_ret, on='scode', how='left')

# 최종 3건 화면 출력
df.tail()
```

```python
def plot_by_factor(df, feat):
    ser = pd.qcut(df[feat], 10, duplicates='drop')  # feat 컬럼별 10개 그룹화 
    print( df.groupby(ser)['RET'].agg(['size', 'mean', 'median']) )  # 그룹별 건수, 평균, 중앙값 출력 
    df.groupby(ser)['RET'].mean().plot(kind='bar') # 그룹별 수익률 평균으로 막대 그래프 생성
    plt.show()

plot_by_factor(df, 'revenue_rate')  # 매출 증가율 
```

```python
plot_by_factor(df, 'oper_income_rate')  # 영업이익 증가율 
```

```python
plot_by_factor(df, 'net_income_rate')  # 순이익증가율
```

```python
plot_by_factor(df, 'ROE')  # ROE(자기자본이익률) 
```
